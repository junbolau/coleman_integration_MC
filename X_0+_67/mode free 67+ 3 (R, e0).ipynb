## This file contains computing $x_i \in \mathbb{Q}$ such that $\omega_0 = \left(\sum_{i=0}^n x_i u^n + O(u^{n+1})\right)d(u)$ for $u=j+j_N$ here by working with $q$-expansions at $R$

#### Fixing a point $S$ on $X^+(67)$, then compute its upper half plan representatives
#### Compute Taylor expansion. I.e., fix a precision $n$, find $x_i \in \mathbb{Q}$ such that $\omega_0 = \left(\sum_{i=0}^n x_i u^n + O(u^{n+1})\right)d(u)$ for $u=j+j_N$ here by working with $q$-expansions.

```python
%%time
#67
mfprec = 15000
M = CuspForms(Gamma0(67),2, prec = mfprec).echelon_form()
A = M.basis()

e0 = 2*A[0]-3*A[1]-3*A[2]+3*A[3]-6*A[4]
e1 = 0*A[0]-A[1]+A[2]+3*A[3]+0*A[4]

cfprec = 5000
CF = ComplexField(cfprec)

EQ = EllipticCurve([1,0,0,-36/(2^6*5^3-1728),-1/(2^6*5^3-1728)])
v = EQ.period_lattice().basis(prec=cfprec)
t1 = v[1]/v[0]
t0 = -1/(-1/t1+2)

a = []
for i in range(67):
    a.append(CF(t0/(i*t0+1)))
b = []
for i in range(0,67):
    b.append(CF(e^(2*pi*I*a[i])))
#q0 = b[19] 
t0 = a[22]
t00 = -1/(67*t0) #this is another representative obtained by W_67 action
q00 = CF(e^(2*pi*I*t00))
t0 = t00
q0 = q00
R.<q> = PowerSeriesRing(QQ,mfprec+100,sparse = True)
omega = sum(e0[i]*q^(i-1) for i in range(mfprec))

matsize = 150

omega_v = [omega(q0)]
omega_d = omega(q)
for i in range(matsize-1):
    omega_d = omega_d.derivative(q)
    omega_v.append(omega_d(q0)/factorial(i+1))
v = vector(omega_v)

etaprec = 1000
eta_1 = 1 + sum((-1)^i*(q^(i*(3*i-1)/2)+q^(i*(3*i+1)/2)) for i in range(1,etaprec))
eta_2 = 1 + sum((-1)^i*(q^(i*(3*i-1))+q^(i*(3*i+1))) for i in range(1,floor(etaprec/2)))


eta_1_taylor = eta_1(q0)
eta_1_dn = eta_1
for i in range(1,matsize):
    eta_1_dn = eta_1_dn.derivative(q)
    eta_1_taylor += eta_1_dn(q0)*(q^i)/factorial(i)
eta_2_taylor = eta_2(q0)
eta_2_dn = eta_2
for i in range(1,matsize):
    eta_2_dn = eta_2_dn.derivative(q)
    eta_2_taylor += eta_2_dn(q0)*(q^i)/factorial(i)
q_taylor = q0 + q

f_taylor = q_taylor*(eta_2_taylor/eta_1_taylor)^24
j_taylor = ((256*f_taylor + 1)^3)/f_taylor
j_d1_taylor = j_taylor.derivative(q)

n = 67

eta_1_n = 1 + sum((-1)^i*(q^(n*i*(3*i-1)/2)+q^(n*i*(3*i+1)/2)) for i in range(1,floor(etaprec/8)))
eta_2_n = 1 + sum((-1)^i*(q^(n*i*(3*i-1))+q^(n*i*(3*i+1))) for i in range(1,floor(etaprec/11)))
q_n = q^67

eta_1_n_taylor = eta_1_n(q0)
eta_1_n_dn = eta_1_n
for i in range(1,matsize):
    eta_1_n_dn = eta_1_n_dn.derivative(q)
    eta_1_n_taylor += eta_1_n_dn(q0)*(q^i)/factorial(i)
eta_2_n_taylor = eta_2_n(q0)
eta_2_n_dn = eta_2_n
for i in range(1,matsize):
    eta_2_n_dn = eta_2_n_dn.derivative(q)
    eta_2_n_taylor += eta_2_n_dn(q0)*(q^i)/factorial(i)
q_n_taylor = q_n(q0)
q_n_dn = q_n
for i in range(1,matsize):
    q_n_dn = q_n_dn.derivative(q)
    q_n_taylor += q_n_dn(q0)*(q^i)/factorial(i)

f_n_taylor = (q_n_taylor)*(eta_2_n_taylor^24)/(eta_1_n_taylor^24)
j_n_taylor = ((256*f_n_taylor + 1)^3)/f_n_taylor
j_n_d1_taylor = j_n_taylor.derivative(q)
```

```python
jinv = j_taylor + j_n_taylor - 2*(8000)
djinv = j_d1_taylor + j_n_d1_taylor
```
